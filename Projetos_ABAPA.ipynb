# ***Tratamento de dados***

```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter

df = pd.read_excel("despesas.xls", skiprows=2, dtype={'Descrição': 'str'})

# Manipulando os dados
df["Mes"] = df["Data"].str[:3]
df["Dia"] = df["Data"].str[4:7]
df["Ano"] = df["Data"].str[7:]
df["Despesas"] = df["Despesas"].str[12:]
df = df.drop(df.index[-1])  # Remover a última linha
df['Valor'] = df['Valor'].astype(float)


saldo = pd.read_excel("saldo.xls", skiprows=2, dtype={'Descrição': 'str'})
saldo = saldo.fillna(0.0)
saldo["Identificação"] = saldo["Identificação"].str[10:]
saldo["Descrição Despesa"] = saldo["Descrição Despesa"].str[12:]
saldo[["Saldo Anterior (+)","Orçado (+)","Realizado (+)","Contrapartida (+)","Remanejamento (+)","Devolução (+)","Remanescente (=)"]].astype(float)
saldo[["Parcela (Projeto)","Ano","Projeto"]] = saldo[["Parcela (Projeto)","Ano","Projeto"]].astype(int)
saldo = saldo.drop(saldo.index[-1])

saldo
```

# **Metricas\:**

```python
# Calcula o total de vendas
total_vendas = df['Valor'].sum()

# Datasets
parcela_total_01   = df.groupby(['Parcelamento','Despesas'])['Valor'].sum().reset_index()
parcela_total_01_1 = df.groupby(['Parcelamento','Atividade'])['Valor'].sum().reset_index()
parcela_total_01_2 = parcela_total_01.pivot(index='Parcelamento', columns='Despesas', values='Valor')
parcela_total_01_3 = df.groupby(['Ano','Mes','Atividade'])['Valor'].sum().reset_index()
parcela_total_01_4 = df.groupby('Parcelamento')['Valor'].sum().reset_index()

saldo_01 = saldo.groupby("Descrição Despesa")['Orçado (+)'].sum().reset_index()

# Métricas
total_vendas_brasil = f"R$ {total_vendas:,.2f}".replace(",", "#").replace(".", ",").replace("#", ".")
qnT_parcelas = df['Parcelamento'].nunique()

# Ajuste para o saldo
saldo_Geral = saldo['Orçado (+)'].sum() - saldo['Realizado (+)'].sum()
saldo_Orcado = saldo['Orçado (+)'].sum()
saldo_Realizado = saldo['Realizado (+)'].sum()
saldo_Contrapartida = saldo['Contrapartida (+)'].sum()
saldo_Remanejamento = saldo['Remanejamento (+)'].sum()
saldo_Devolucao = saldo['Devolução (+)'].sum()
saldo_Remanescente = saldo['Remanescente (=)'].sum()

# Gasto com gasolina
#Gasolina = saldo.query("`Descrição Despesa` == 'Gasolina'")['Realizado (+)'].sum()
#Gasolina2 = parcela_total_01.query("`Despesas` == 'Gasolina'")['Valor'].sum()

# Exibição dos resultados
print(f"Total de gasto: {total_vendas_brasil}")
print(f"Quantidade de parcelas: {qnT_parcelas}")

print(f"---------------------------------------")
print(f"Orçado: R$ {saldo_Orcado:,.2f}")
print(f"Realizado: R$ {saldo_Realizado:,.2f}")
print(f"Contrapartida: R$ {saldo_Contrapartida:,.2f}")
print(f"Devolução: R$ {saldo_Devolucao:,.2f}")
print(f"Remanescente: R$ {saldo_Remanescente:,.2f}")
print(f"Saldo: R$ {saldo_Geral:,.2f}")

saldo_01
```
